```python
from selenium import webdriver

path ='E:\chromedriver.exe' 
driver = webdriver.Chrome(path)

driver.get("https://www.myntra.com/shoes")
```

```python
price_tags=driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div')
price_tags
```

```python
price_tags.click()
```

```python
color_tags=driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div')
color_tags
```

```python
color_tags.click()
```

```python
brand=driver.find_elements_by_xpath('//h3[@class="product-brand"]')
len(brand)
```

```python
des=driver.find_elements_by_xpath('//h4[@class="product-product"]')
len(des)
```

```python
price=driver.find_elements_by_xpath('//div[@class="product-price"]')
len(price)
```

```python
Brand=[]

for i in brand:
    Brand.append(i.text)
    
len(Brand)
```

```python
Desc=[]

for i in des:
    Desc.append(i.text)
    
len(Desc)
```

```python
Price=[]

for i in price:
    Price.append(i.text)
    
len(Price)
```

```python
df=pd.DataFrame({'Brand':Brand,'Short_Description':Desc,'Product_Price':Price})
df
```
