```python
page=requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)

soup
```

```python
movie_details = soup.find_all('div',class_="lister")

name1 = []

for store in movie_details:
    name = store.a.find_all('href')
    name1.append(name)

movie_name=[]

for i in soup.find_all('td',class_="titleColumn"):
    movie_name.append(i.text.replace('\n','',))
    movie_name
    
rating = []

for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text.replace('\n',''))
    rating

year_of_release = []

for i in soup.find_all('span',class_="secondaryInfo"):
    year_of_release.append(i.text)
    year_of_release
```

```python
df = pd.DataFrame({'Movie_Name':movie_name,'Rating':rating,'Year_Of-Release':year_of_release})
```
